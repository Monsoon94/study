## 5의 배수인 행들만 인덱싱하기 ( 0번째 행 미포함)

```python
def multiples(value, length):
    return [value * i for i in range(1, length+1)]

mul5 = multiples(5, 500)
print(mul5)
```

```python
df.loc[df.index.isin(mul5)]
```
