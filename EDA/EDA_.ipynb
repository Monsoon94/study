## 결측치 확인

### 결측치 개수 확인

```python
import pandas as pd
import numpy as np

df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/titanic/train.csv')
df.head()
```

```python
df.isna().sum()
```

### 결측치 비율로 표현하기

```python
df.isnull().mean() * 100
```

```python
df1 = df.isnull().mean() * 100
df2 = df1.sort_values(ascending=False)
df2
```

```python
percent_missing = df.isnull().sum() * 100 / len(df)
missing_value_df = pd.DataFrame({'column_name': df.columns,
                                 'percent_missing': percent_missing})

missing_value_df_sorted = missing_value_df.sort_values('percent_missing', ascending=False)
missing_value_df_sorted
```

### 변수 별 결측치 확인

```python
# 결측치를 포함하여 빈도 확인
df['Cabin'].value_counts(dropna=False) #ascending=True가 기본값.
```

```python
# 각 factor의 비율을 출력
df['Cabin'].value_counts(dropna=False, normalize=True) #ascending=True가 기본값.
```

### 결측치에 해당하는 행 출력

```python
df.loc[df['Age'].isna()]
```
